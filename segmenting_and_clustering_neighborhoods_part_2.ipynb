## Session 1: APIs test for data fetching

```python
def lat_lng_func(postal_code):
    
    geolocator = Nominatim(user_agent="canada_explorer")
    address = '{}, Toronto, Ontario'.format(postal_code)

    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    
    return latitude,longitude
```

```python
def lat_lng_func_02(postal_code):
    
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
        lat_lng_coords = g.latlng
        print (lat_lng_coords)

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]
    
    return latitude,longitude
```

## Session 2: coordinate data usage by .csv file

```python
df_coordinate = pd.read_csv('Geospatial_Coordinates.csv')
df_coordinate.head()
```

```python
df_coordinate.rename(columns = {'Postal Code':'Postal_Code'},inplace=True)
df_coordinate.head()
```

```python
df_process = df.copy()
df_process['Latitude'] = 0
df_process['Longitude'] = 0
```

```python
for row in range(df_process.shape[0]):
    
    postal_code = df_process.iloc[row,0]
    
    try:
        lat = df_coordinate[df_coordinate.Postal_Code == postal_code]['Latitude'].values[0]
        lon = df_coordinate[df_coordinate.Postal_Code == postal_code]['Longitude'].values[0]
    
    except:
        
        lat = np.NaN
        lon = np.NaN
        
    df_process.iloc[row,3] = lat
    df_process.iloc[row,4] = lon

df_process.to_excel('segmenting_and_clustering_part_2.xlsx')  
df_process    
```
